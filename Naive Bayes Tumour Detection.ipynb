```python
benign = pd.DataFrame(rawdata[rawdata['diagnosis']=='B'])
```

```python
benign_training_data = benign.iloc[0:199]
benign_training_data = benign_training_data['radius_mean']
```

```python
benign_training_data
```

```python
benign_testing_data = benign.iloc[199:]
benign_testing_data = pd.DataFrame(benign_testing_data['radius_mean'])
```

```python
benign_testing_data
```

```python
mu_cap_m_radius_mean = malignant_training_data.mean()
```

```python
sigma_cap_m_radius_mean = malignant_training_data.std()
```

```python
mu_cap_b_radius_mean = benign_training_data.mean()
```

```python
sigma_cap_b_radius_mean = benign_training_data.std()
```

```python
precision_sum = 0
recall_sum = 0 
accuracy_sum = 0 

for i in range(0,len(malignant_testing_data)):
    class_conditional_benign = s.norm.pdf(malignant_testing_data.iloc[i],mu_cap_b_radius_mean,sigma_cap_b_radius_mean)
    
    class_conditional_malignant = s.norm.pdf(malignant_testing_data.iloc[i],mu_cap_m_radius_mean,sigma_cap_m_radius_mean)
    
    posterior_benign_prob = (class_conditional_benign)/(class_conditional_benign + class_conditional_malignant)
    
    posterior_malignant_prob = 1 - posterior_benign_prob
    
    if posterior_benign_prob < 0.5:
        accuracy_sum += 1
```

```python
accuracy_sum
```

```python
for i in range(0,len(benign_testing_data)):
    class_conditional_benign = s.norm.pdf(benign_testing_data.iloc[i],mu_cap_b_radius_mean,sigma_cap_b_radius_mean)
    
    class_conditional_malignant = s.norm.pdf(benign_testing_data.iloc[i],mu_cap_m_radius_mean,sigma_cap_m_radius_mean)
    
    posterior_benign_prob = (class_conditional_benign)/(class_conditional_benign + class_conditional_malignant)
    
    posterior_malignant_prob = 1 - posterior_benign_prob
    
    if posterior_benign_prob > 0.5:
        accuracy_sum += 1
```
